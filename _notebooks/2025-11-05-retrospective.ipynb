## Big Idea 3 Reflection

### 3.1 ‚Äì Variables  
- Variables store information for later use in a program.  
- I learned how to assign values correctly.  
- Using clear, descriptive names makes code easier to read and maintain.

### 3.2 ‚Äì Lists and Strings  
- Lists help manage complexity by storing many related items in one variable.  
- They can grow or shrink as needed and allow the same computation on all elements.  
- Strings and lists are both ordered sequences.

### 3.3 ‚Äì Algorithms and Teaching  
- I created JavaScript and Python homework.  
- Learned about learning objectives and algorithms.  
- Taught others how algorithms work.

### 3.4 ‚Äì Strings  
- Strings are sequences of characters inside quotes used to store text.  
- I learned to define strings in both Python and JavaScript.  
- Practiced using operations like upper, lower, slicing, and concatenation.  
- Learned how to create multiline strings.

### 3.5 ‚Äì Boolean Expressions  
- Booleans help programs make decisions with True or False values.  
- Relational operators compare values, and modulus checks remainders.  
- Logical operators like not, and, or combine or flip conditions.

### 3.6 ‚Äì Conditionals  
- Conditionals let programs make decisions based on conditions.  
- I used if, elif/else if, and else to control code flow.  
- Learned to use comparison and logical operators with Booleans.

### 3.7 ‚Äì Nested Conditionals  
- Nested conditionals are if statements inside other if statements.  
- They help handle multiple related decisions in order.  
- Useful for complex logic like age checks or game conditions.

### 3.8 ‚Äì Iteration  
- Iterations repeat code using loops.  
- For loops go through sequences, while loops repeat while a condition is True.  
- Loops can be nested or indexed with enumerate().

### 3.9 ‚Äì Developing Algorithms  
- Algorithms are step-by-step instructions to solve problems.  
- Clear algorithms make code easier to understand, test, and debug.  
- Efficiency matters when designing algorithms.

### 3.10 ‚Äì Lists  
- Lists organize and process multiple pieces of data.  
- I practiced adding, removing, and looping through list elements.  
- Lists make programs more flexible and dynamic.

### 3.12 ‚Äì Calling Procedures  
- Calling procedures helps reuse code efficiently.  
- I learned to pass arguments into functions.  
- Functions simplify and organize complex programs.

### 3.13 ‚Äì Developing Procedures  
- I learned how to define my own procedures.  
- Functions make programs cleaner and more organized.  
- Reusable code reduces errors and improves structure.

### 3.15 ‚Äì Random Values  
- Random values add variety and unpredictability to programs.  
- Useful in games, simulations, and testing.  
- I learned how to generate and use random numbers in code.

### 3.17 ‚Äì Algorithmic Efficiency  
- Algorithmic efficiency measures how fast and resourceful a program is.  
- Efficient algorithms save time and improve performance.  
- I learned to compare different approaches to find the best one.
